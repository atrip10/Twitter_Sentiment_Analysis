```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
w = "better"
pos_tag([w])
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
stops, string.punctuation
```

```python
x_train
x_train_new = []
for sentence in x_train:
    x_train_new.append(sentence.split())
x_train = x_train_new
x_test_new = []
for sentence in x_test:
    x_test_new.append(sentence.split())
x_test = x_test_new
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            if(clean_word[0] == '@'):
                clean_word_new = clean_word[1:]
                clean_word = clean_word_new
            output_words.append(clean_word.lower())
    return output_words
```

```python
x_train_updated = [clean_review(x) for x in x_train]
x_test_updated = [clean_review(x) for x in x_test]
```
